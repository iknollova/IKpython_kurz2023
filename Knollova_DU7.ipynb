Řazení se u speciálních znaků a české diakritiky, použítí unidecode (viz TIP). Ale úplně si nejsem jistá, že by to bylo to, zde se mi jen odstranily speciální české znaky ze žluvy je zluva. úplně správně by mělo být řazení dle české abecedy se speciálními znaky tj. c, č..., z,ž. 

```python
from unidecode import unidecode

adopce_zvirat["nazev_cz"]=adopce_zvirat["nazev_cz"].apply(unidecode)
```

```python
adopce_zvirat.sort_values(by=["nazev_cz"],ascending= False).head(30)
```

```python
adopce_zvirat.sort_values(by=["nazev_cz"],ascending= False).head(1)
```
